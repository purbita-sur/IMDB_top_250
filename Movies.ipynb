# Importing libraries

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url_list=['https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc&ref_=adv_prv','https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc&start=51&ref_=adv_nxt','https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc&start=101&ref_=adv_nxt','https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc&start=151&ref_=adv_nxt','https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc&start=201&ref_=adv_nxt']
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.1; Win64; x64)','Accept-Language': 'en-US,en;q=0.8'}
```

# 1. Movie Id

```python
movie_id_elem=[]
movie_id=[]
for url in url_list:
    page=requests.get(url,headers=headers)
    soup=BeautifulSoup(page.text,'html.parser')
    id_temp=soup.find_all(class_='lister-item-index unbold text-primary')
    movie_id_elem=movie_id_elem+id_temp
for item in movie_id_elem:
    s=item.text
    s=s.strip('.')
    movie_id.append(s)
len(movie_id)
```

```python
movie_id
```

# 2. Movie Name

```python
movie_name_elem=[]
movie_name=[]
for url in url_list:
    page=requests.get(url,headers=headers)
    soup=BeautifulSoup(page.text,'html.parser')
    name_temp=soup.find_all(class_='lister-item-header')
    movie_name_elem=movie_name_elem+name_temp
for item in movie_name_elem:
    movie_name.append((item.find('a')).text)
```

```python
len(movie_name)
```

# 3. Year of Release

```python
movie_year_elem=[]
movie_year=[]
for url in url_list:
    page=requests.get(url,headers=headers)
    soup=BeautifulSoup(page.text,'html.parser')
    year_temp=soup.find_all(class_='lister-item-year text-muted unbold')
    movie_year_elem=movie_year_elem+year_temp
for item in movie_year_elem:
    s=item.text
    s=s.strip('(')
    s=s.strip(')')
    movie_year.append(s)
movie_year
```

```python
len(movie_year)
```

# 4. Genre of Movie

```python
movie_genre_elem=[]
movie_genre=[]
for url in url_list:
    page=requests.get(url,headers=headers)
    soup=BeautifulSoup(page.text,'html.parser')
    genre_temp=soup.find_all(class_='genre')
    movie_genre_elem=movie_genre_elem+genre_temp
for item in movie_genre_elem:
    s=item.text
    s=s.strip('\n')
    s=s.strip()
    movie_genre.append(s)
movie_genre
```

```python
len(movie_genre)
```

# 5. Imdb Rating

```python
movie_rating=[]
for url in url_list:
    page=requests.get(url,headers=headers)
    soup=BeautifulSoup(page.text,'html.parser')
    rating=soup.find_all('div',{'class':'inline-block ratings-imdb-rating'})

    for item in rating:
        movie_rating.append(item.find('strong').text)
movie_rating
```

```python
len(movie_rating)
```

# 6. Movie Metascore

```python
movie_score_elem=[]
for url in url_list:
    page=requests.get(url,headers=headers)
    soup=BeautifulSoup(page.text,'html.parser')
    score_temp=soup.find_all('div',{'class':'ratings-bar'})
    movie_score_elem=movie_score_elem+score_temp
```

```python
movie_score=[]
for i in movie_score_elem:
    try:
        s=(i.find('div',{'class':'inline-block ratings-metascore'}))
        s=s.find('span',{'class':'metascore favorable'})
        s=s.text
        s=s.strip()
        movie_score.append(s)
    except:
        movie_score.append('NaN')
```

```python
len(movie_score)
```

```python
movie_score
```

# 7. Movie Summary

```python
movie_summary_elem=[]
movie_summary=[]
for url in url_list:
    page=requests.get(url,headers=headers)
    soup=BeautifulSoup(page.text,'html.parser')
    summary_temp=soup.find_all('p',{'class':'text-muted'})
    for item in summary_temp:
        movie_summary_elem.append(item.text)
for i in range(len(movie_summary_elem)):
    if i%2!=0:
        s=movie_summary_elem[i]
        s=s.strip('\n')
        s=s.strip()
        movie_summary.append(s)
movie_summary
```

```python
len(movie_summary)
```

# 8. Director

```python
movie_director_elem=[]
movie_director_temp=[]
movie_director=[]
for url in url_list:
    page=requests.get(url,headers=headers)
    soup=BeautifulSoup(page.text,'html.parser')
    director_temp=soup.find_all('p',{'class':""})
    movie_director_elem=movie_director_elem+director_temp

for item in movie_director_elem:
    movie_director_temp.append(item.find('a'))
s=2
e=52
for j in range(0,5):
    for i in range(s,e):
        movie_director.append(movie_director_temp[i].text)
    s+=53
    e+=53
```

# 9. Movie Stars

```python
movie_stars_elem=[]
movie_stars_temp=[]
movie_stars=[]
for url in url_list:
    page=requests.get(url,headers=headers)
    soup=BeautifulSoup(page.text,'html.parser')
    stars_temp=soup.find_all('p',{'class':""})
    movie_stars_elem=movie_stars_elem+stars_temp
for item in movie_stars_elem:
    movie_stars_temp.append(item.find_all('a'))
```

```python
s=2
e=52
for j in range(0,5):
    for i in range(s,e):
        movie_stars.append(movie_stars_temp[i][1:])
    s+=53
    e+=53
```

```python
stars=[]
m=''
for i in range(0,250):
    for j in movie_stars[i]:
        m=m+j.text+','
    stars.append(m)
    m=''
```

# 10. Number Of Votes

```python
movie_votes_elem=[]
movie_votes=[]
for url in url_list:
    page=requests.get(url,headers=headers)
    soup=BeautifulSoup(page.text,'html.parser')
    votes_temp=soup.find_all('p',{'class':'sort-num_votes-visible'})
    movie_votes_elem=movie_votes_elem+votes_temp
for item in movie_votes_elem:
    movie_votes.append(item.find('span',{'name':'nv'}).text)
```

```python
movie_votes
```

```python
len(movie_votes)
```

# 11. Gross Value

```python
movie_gross_elem=[]
movie_gross_temp=[]
movie_gross_1=[]
movie_gross=[]
for url in url_list:
    page=requests.get(url,headers=headers)
    soup=BeautifulSoup(page.text,'html.parser')
    gross_temp=soup.find_all('p',{'class':'sort-num_votes-visible'})
    movie_gross_elem=movie_gross_elem+gross_temp
for item in movie_gross_elem:
    movie_gross_temp.append(item.find_all('span',{'name':'nv'}))
```

```python
for i in range(0,len(movie_gross_temp)):
    try:
        movie_gross.append(movie_gross_temp[i][1].text)
    except:
        movie_gross.append('NaN')
```

```python
len(movie_gross)
```

# Creating the Dataframe

```python
movies={}
movies['Id']=movie_id
movies['Name']=movie_name
movies['Year of Release']=movie_year
movies['Genre']=movie_genre
movies['IMDB Rating']=movie_rating
movies['Metascore']=movie_score
movies['Director']=movie_director
movies['Summary']=movie_summary
movies['Stars']=stars
movies['Votes']=movie_votes
movies['Gross Income']=movie_gross
movie_df = pd.DataFrame(data=movies)
movie_df.to_csv('IMDB_Movies.csv')
```
